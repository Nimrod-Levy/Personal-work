**Food testing on 300 ingridients:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import tree

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('../input/300-fooods/300 foods utf 8 .csv')

data
```

```python
data = data.drop(['name','IS KETO','serving_size'] , axis = 1)
y = data['is it KETO']
X = data.drop('is it KETO', axis=1)
data
```

```python

scaler = StandardScaler()


X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X = X.fillna(0)

X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
log_model = LogisticRegression()
svm_model = SVC(C=1.0, kernel='rbf')
nn_model = MLPClassifier(hidden_layer_sizes=(256, 256))
tree = tree.DecisionTreeClassifier()
```

```python
np.sum(y) / len(y)
```

```python
log_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
clf = nn_model.fit(X_train, y_train)
tree.fit(X_train, y_train)
```

```python
data.insert(1, "newcol1", clf.predict(X))
data
```

```python
print(f"---Logistic Regression: {log_model.score(X_test, y_test)}")
print(f"Support Vector Machine: {svm_model.score(X_test, y_test)}")
print(f"--------Neural Network: {nn_model.score(X_test, y_test)}")
print(f"------------------tree: {tree.score(X_test, y_test)}")

```
